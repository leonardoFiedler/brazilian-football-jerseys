```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "../../data/processed/brazil-teams-jersey-price-processed.csv",
    usecols=[
        "team",
        "state",
        "year",
        "value",
        "value_int",
        "value_add_05",
        "value_add_10",
        "value_add_15",
        "value_sub_05",
        "value_sub_10",
        "value_sub_15",
    ],
)

df.head()
```

```python
flamengo_df = df.query("team == 'Flamengo'")

plt.figure(figsize=(16, 6))

g = sns.relplot(
    data=flamengo_df, kind="line", x="year", y="value_int", hue="team", marker="o"
)
plt.xticks(
    flamengo_df["year"], rotation=45
)  # Explicitly set the ticks to every year in the dataframe

plt.title("Evoluo preo da camisa oficial do Flamengo (2012/24)", fontsize=15)
plt.xlabel("Anos", fontsize=14)
plt.ylabel("Valores (R$)", fontsize=14)

sns.move_legend(g, "upper left", bbox_to_anchor=(0.90, 0.45), title="Time")

plt.tight_layout()
plt.show()
```
